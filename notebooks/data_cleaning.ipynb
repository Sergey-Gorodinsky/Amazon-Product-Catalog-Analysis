# Checking and cleaning the data using Pandas

### 1. Remove redundant columns;

```python
df.columns
```

` 'img_link', 'product_link' ` columns are not required for analysis as they do not provide any usefull information or can be used to extract some key insights/information

```python
df.drop(columns=['img_link', 'product_link'], inplace=True)
```

### 3. Remove duplicates;

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

### 4. Check data types;

```python
df.dtypes
```

* discounted_price       
* actual_price           
* discount_percentage    
* rating                 
* rating_count 
          
Those are numerical values and it would be easier to work with them if the had the appropriate data types

```python
df[['discounted_price',
       'actual_price', 'discount_percentage', 'rating', 'rating_count']]
```

(1) Get rid of ₹ and % symbols

```python

df["discounted_price"] = df["discounted_price"].str.replace("₹","").replace(" ","")

df["actual_price"] = df["actual_price"].str.replace("₹","").replace(" ","")

df[["discounted_price", "actual_price"]]
```

```python
df["discount_percentage"] = df["discount_percentage"].str.replace("%","").replace(" ", "")

df["discount_percentage"]
```

```python
df[['discounted_price','actual_price', 
    'discount_percentage', 'rating', 'rating_count']]
```

(2) Removing commas

```python
df["discounted_price"] = df["discounted_price"].str.replace(",","")
df["actual_price"] = df["actual_price"].str.replace(",","")
df["rating_count"] = df["rating_count"].str.replace(",","")

df[['discounted_price','actual_price', 
    'discount_percentage', 'rating', 'rating_count']]
```

(3) Changing the data types

```python
df["rating"] = df["rating"].str.replace("|","")

df["rating"] = pd.to_numeric(df["rating"])
```

```python
numeric_data = ["discounted_price", "actual_price", "rating_count"]

for item in numeric_data:
    df[item] = pd.to_numeric(df[item])

df["discount_percentage"] = df["discount_percentage"].astype(int)
```

```python
df[['discounted_price',
       'actual_price', 'discount_percentage', 'rating', 'rating_count']]
```

```python
df[['discounted_price',
       'actual_price', 'discount_percentage', 'rating', 'rating_count']].dtypes
```

```python
for col in df.select_dtypes(include=["object","string"]):
    df[col] = df[col].astype(str).str.encode(encoding="ascii",errors="ignore").str.decode(encoding="ascii")
```

```python
df.head(20)
```

### 5. Check NaN values;

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

### 6. Check the dataset for logical errors;

```python
df.columns
```

```python
# checking if numeric values are positive

numeric = df[['discounted_price',
       'actual_price', 'discount_percentage', 'rating', 'rating_count']]

for value in numeric:
    print(f"{value} Min -> {df[value].min()}")
```

### 7. Identify outliers;

```python
numeric = df[['discounted_price','actual_price', 
    'discount_percentage', 'rating', 'rating_count']]

# fig, axes = plt.subplots(5,2, figsize=(20,16))

for col in numeric:
    plt.figure(figsize=(15,7))
    sns.boxplot(data=df,x=col,color="#6C00E0")

plt.show()
```
